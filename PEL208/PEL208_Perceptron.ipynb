```python
import pandas as pd

iris_2_data = pd.read_csv("/home/pi/PEL208/data/iris_2_classes.txt", usecols=[0, 1, 2, 3], header=None)
iris_2_label = pd.read_csv("/home/pi/PEL208/data/iris_2_classes.txt", usecols=[4], header=None)
classes = {'setosa': 1, 'versicolor': 0}
iris_2_label[4] = [classes[item] for item in iris_2_label[4]]
iris_2_label.head()
```

```python
import numpy as np

n = 1

def fit(x, t):
    cols = len(x)
    rows = len(t)
    x = x.to_numpy()
    t = t.to_numpy()
    w = np.zeros((rows, cols))
    
    y = perceptron_output(w, 1, x)
    dw = n * x * (t - y)
    w = w + dw
    
    return w

def step_function(x):
    return 1 if x >= 0 else 0

def perceptron_output(weights, bias, x):
    calculation = np.dot(weights, x) + bias
    return step_function(calculation)

result = fit(iris_2_data, iris_2_label)
result
```
