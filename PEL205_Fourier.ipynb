# PEL205 - Fourier

```python
import numpy as np
from scipy.fft import fft, fft2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```python
img = mpimg.imread("./data/images/ecovias-001.jpeg")
# img = mpimg.imread("./data/images/fourier.png")
img_bw = rgb2gray(img)
```

```python
img2 = fft(img_bw)
```

```python
f = plt.figure(figsize=[18, 13.5])
f.add_subplot(2, 2, 1)
plt.imshow(img)
f.add_subplot(2, 2, 2)
plt.imshow(img_bw, cmap=plt.get_cmap("gray"))
f.add_subplot(2, 2, 3)
plt.imshow(np.log(1+np.abs(img2)), cmap=plt.get_cmap("gray"))
plt.show()
```
